```python
import pandas as pd

# Lê o arquivo com o nome correto (dupla extensão .csv.csv)
df = pd.read_csv("Superstore.csv.csv", encoding="latin1")

# Visualiza as primeiras linhas
df.head()
```

```python
# Informações gerais do dataset
df.info()

# Verificar valores nulos
df.isnull().sum()

# Estatísticas descritivas
df.describe(include='all')
```

```python
# Remover duplicatas
df.drop_duplicates(inplace=True)

# Conferir novamente
df.info()
```

```python
# Total de vendas por categoria
df.groupby("Category")["Sales"].sum().sort_values(ascending=False).plot(kind='bar', title="Total de Vendas por Categoria")
plt.ylabel("Vendas")
plt.show()

# Lucro por subcategoria
df.groupby("Sub-Category")["Profit"].sum().sort_values().plot(kind='barh', title="Lucro por Subcategoria")
plt.xlabel("Lucro")
plt.show()

# Vendas ao longo do tempo
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.set_index("Order Date")["Sales"].resample("M").sum().plot(title="Vendas Mensais")
plt.ylabel("Vendas")
plt.show()
```

```python
#Top 10 produtos mais lucrativos

df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
#Produtos que geraram prejuízo

df.groupby("Product Name")["Profit"].sum().sort_values().head(10)
```

```python
#Vendas por Macroregião

df.groupby("Region")["Sales"].sum().sort_values(ascending=False).plot(kind="bar", title="Vendas por Região")
```

Tradução para Português

```python
#Nomes das colunas

df.columns
```

```python
#Código para aplicar as traduções

df = df.rename(columns={
    "Row ID": "ID da Linha",
    "Order ID": "ID do Pedido",
    "Order Date": "Data do Pedido",
    "Ship Date": "Data de Envio",
    "Ship Mode": "Modalidade de Envio",
    "Customer ID": "ID do Cliente",
    "Customer Name": "Nome do Cliente",
    "Segment": "Segmento",
    "Country": "País",
    "City": "Cidade",
    "State": "Estado",
    "Postal Code": "CEP",
    "Region": "Região",
    "Product ID": "ID do Produto",
    "Category": "Categoria",
    "Sub-Category": "Subcategoria",
    "Product Name": "Nome do Produto",
    "Sales": "Vendas",
    "Quantity": "Quantidade",
    "Discount": "Desconto",
    "Profit": "Lucro"
})
```

```python
print(df["Categoria"].unique())
print(df["Subcategoria"].unique())
print(df["Segmento"].unique())
print(df["Região"].unique())
```

```python
#Traduzir valores da coluna categoria

df["Categoria"] = df["Categoria"].replace({
    "Furniture": "Móveis",
    "Office Supplies": "Material de Escritório",
    "Technology": "Tecnologia"
})
```

```python
#Traduzir valores da coluna Subcategoria

df["Subcategoria"] = df["Subcategoria"].replace({
    "Bookcases": "Estantes",
    "Chairs": "Cadeiras",
    "Labels": "Etiquetas",
    "Tables": "Mesas",
    "Storage": "Armazenamento",
    "Furnishings": "Mobília",
    "Art": "Arte",
    "Phones": "Telefones",
    "Binders": "Pastas",
    "Appliances": "Eletrodomésticos",
    "Paper": "Papel",
    "Accessories": "Acessórios",
    "Envelopes": "Envelopes",
    "Fasteners": "Fixadores",
    "Supplies": "Suprimentos",
    "Machines": "Máquinas",
    "Copiers": "Copiadoras"
})
```

```python
#Traduzir valores da coluna Segmento

df["Segmento"] = df["Segmento"].replace({
    "Consumer": "Consumidor",
    "Corporate": "Empresarial",
    "Home Office": "Escritório em Casa"
})
```

```python
#Traduzir valores da coluna Região

df["Região"] = df["Região"].replace({
    "South": "Sul",
    "West": "Oeste",
    "Central": "Central",
    "East": "Leste"
})
```

Gerando os gráficos em Português

```python
#Gráfico 1: Lucro por Subcategoria

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
ordem = df.groupby("Subcategoria")["Lucro"].sum().sort_values().index
sns.barplot(x="Lucro", y="Subcategoria", data=df, order=ordem, palette="crest")

plt.title("Lucro por Subcategoria")
plt.xlabel("Lucro Total")
plt.ylabel("Subcategoria")
plt.tight_layout()
plt.show()
```

```python
#Gráfico 2: Vendas Mensais

df["Data do Pedido"] = pd.to_datetime(df["Data do Pedido"])
df["AnoMes"] = df["Data do Pedido"].dt.to_period("M")

vendas_mensais = df.groupby("AnoMes")["Vendas"].sum().reset_index()
vendas_mensais["AnoMes"] = vendas_mensais["AnoMes"].astype(str)

plt.figure(figsize=(14, 6))
sns.lineplot(data=vendas_mensais, x="AnoMes", y="Vendas", marker="o")

plt.title("Vendas Mensais")
plt.xlabel("Ano-Mês")
plt.ylabel("Vendas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Gráfico 3: Vendas por Região

plt.figure(figsize=(8, 6))
ordem = df.groupby("Região")["Vendas"].sum().sort_values(ascending=False).index
sns.barplot(x="Região", y="Vendas", data=df, order=ordem, palette="viridis")

plt.title("Vendas por Região")
plt.xlabel("Região")
plt.ylabel("Total de Vendas")
plt.tight_layout()
plt.show()
```
